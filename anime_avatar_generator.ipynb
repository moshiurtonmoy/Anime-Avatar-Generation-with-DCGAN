```python
import tensorflow as tf
from tensorflow.keras.utils import array_to_img
import matplotlib.pyplot as plt
```

```python
generator = tf.keras.models.load_model('/content/generator.h5')
```

```python
LATENT_DIM = 128

def generate_new_img(num_imgs=50):
    noise = tf.random.normal([num_imgs, LATENT_DIM]) # random noise
    gen_img = generator(noise) # generate new img from random_noise
    gen_img = (gen_img * 127.5) + 127.5 # denormalizing the generated img
    gen_img = gen_img.numpy()

    # plot the generated img
    fig = plt.figure(figsize=(14,7)) # displaying generated img samples after on each epoch
    for i in range(num_imgs):
        plt.subplot(5, 10, i+1)
        img = array_to_img(gen_img[i])
        plt.imshow(img)
        plt.axis('off')
    plt.subplots_adjust(wspace=0.05, hspace=0.05)
    plt.show()
```

```python
generate_new_img()
```
